```python
import pandas as pd
import os
```

```python
current_directory = os.getcwd()
file_name = "bearing_lock_nuts_square.xlsx"
file_path = os.path.abspath(os.path.join(current_directory,"..","data",file_name))
```

```python
table = pd.read_excel(file_path)
```

```python
table
```

```python
table = table.replace("-","none")
```

```python
for_payload = table.drop(columns=["Part Number URL"])
```

```python
for_payload
```

```python
table["row-text"] = table.apply(lambda r:" | ".join(f"{c}: {r[c]}" for c in for_payload.columns),axis=1)
```

```python
table["row-text"][0]
```

```python
not_for_embedding = {
    "Price", "Volumn Discount", "Minimum order Qty",
    "Days to Ship", "Seal", "Part Number URL","Volume Discount","Part Number URL","row-text"
}
```

```python
table["embedding"] = table.apply(lambda r: " | ".join(f"{c}: {r[c]}" for c in for_payload.columns if c not in not_for_embedding), axis = 1)
```

```python
table["embedding"][0]
```

```python
for_specification_exclusion = {
    "Part Number Name", "embedding", "row-text", "Price", "Volumn Discount", "Minimum order Qty",
    "Days to Ship", "Seal", "Part Number URL","Volume Discount","Part Number URL"
}
```

```python
table["specification"] = table.apply(lambda r: " | ".join(f"{c}: {r[c]}" for c in for_payload.columns if c not in for_specification_exclusion), axis = 1)
```

```python
table["specification"][0]
```
