```python
from langchain_text_splitters import MarkdownHeaderTextSplitter
```

```python
headers_to_split_on = [("#","Product")]
```

```python
markdown_documents = MarkdownHeaderTextSplitter(
    headers_to_split_on=headers_to_split_on
)
```

```python
with open("all_mobilebigbyte_feature.md","r",encoding = "utf-8") as f:
    markdown_text = f.read()
```

```python
product_docs = markdown_documents.split_text(markdown_text)
```
